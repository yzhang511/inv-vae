```python
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
small_motion = np.c_[np.random.normal(0, 1, 20), np.random.normal(0, 1, 20)]
large_motion = np.c_[np.random.normal(2, 1, 30), np.random.normal(2, 1, 30)]
```

```python
fig, ax = plt.subplots(1,1, figsize=(3,3))
plt.scatter(small_motion.T[0], small_motion.T[1], c="royalblue", edgecolor="white", s=200, alpha=1., linewidth=2)
plt.scatter(large_motion.T[0], large_motion.T[1], c="darkorange", edgecolor="white", s=200, alpha=1., linewidth=2);
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.set_xticks([], [])
ax.set_yticks([], [])
```
